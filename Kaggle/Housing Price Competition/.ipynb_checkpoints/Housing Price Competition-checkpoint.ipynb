#### 주거용 주택의 모든 측면을 설명하는 79개의 변수를 활용하여 각 주택의 최종 가격을 예측
##### 평가 산식 : RMSE

#### 데이터 변수 설명

```python
# SalePrice : 주택의 판매 가격
# MSSubClass: 주택 유형
# MSZoning: 일반 구역 분류
# LotFrontage: 주택에 연결된 거리의 선형 피트
# LotArea: 주택 평방 피트 단위의 부지 크기
# Street: 도로 접근 유형
# Alley: 골목 접근 유형
# LotShape: 주택의 일반적인 모양
# LandContour: 주택의 평탄도
# Utilities: 사용가능한 유틸리티 유형
# LotConfig: Lot의 구성
# LandSlope: 속성의 기울기
# Neighborhood: Ames시 이내 위치
# Condition1: 주요 도로 또는 철도와 인접
# Condition2: 주요 도로 또는 철도와 인접(초가 있는 경우)
# BldgType: 주거 유형
# HouseStyle: 주거 스타일
# OverallQual: 전체 재료 및 마감 품질
# OverallCond: 전체 상태 등급
# YearBuilt: 건설 날짜
# YearRemodAdd: 리모델링 날짜
# RoofStyle: 지붕 유형
# RoofMatl: 지붕 재료
# Exterior1st: 집의 외부 덮개
# Exterior2nd:  주택의 외부 덮개(두 개 이상의 재료인 경우)
# MasVnrType: 석조 베니어 유형
# MasVnrArea: 석조 베니어판 면적(제곱피트)
# ExterQual: 외장재 품질
# ExterCond: 외장재의 상태 표시
# Foundation: 기초(기반)의 종류
# BsmtQual: 지하실 높이
# BsmtCond: 지하의 일반적인 상태
# BsmtExposure: 작업 중단 또는 정원 수준의 지하 벽
# BsmtFinType1: 지하 마감 구역 품질
# BsmtFinSF1: 1종 완제품 평방 피트 
# BsmtFinType2: 두 번째 마감된 영역의 품질(존재 할 때)
# BsmtFinSF2: 2종 완제품 평방 피트
# BsmtUnfSF: 지하 1평방피트 미완성
# TotalBsmtSF: 지하 총 면적 평방 피트
# Heating: 난방 종류
# HeatingQC: 난방 품질 및 상태
# CentralAir: 중앙 에어컨
# Electrical: 전기 시스템
# 1stFlrSF: 1층 평방 피트
# 2ndFlrSF: 2층 평방 피트
# LowQualFinSF: 저품질 마감 평방 피트(모든 층)
# GrLivArea: 지상 생활 공간 평방 피트
# BsmtFullBath: 지하욕실이 가득 찬 욕실
# BsmtHalfBath: 지하욕실이 반만 찬 욕실
# FullBath: 지상욕실이 가득 찬 욕실
# HalfBath: 지상욕실이 반만 찬 욕실
# Bedroom: 지하 층 이상의 침실 수
# Kitchen: 주방 개수
# KitchenQual: 주방 품질
# TotRmsAbvGrd: 지상 총 객실(화장실 미포함)
# Functional: 가정용 기능 등급
# Fireplaces: 벽난로 수
# FireplaceQu: 벽난로 품질
# GarageType: 차고 위치
# GarageYrBlt: 차고가 지어진 년도
# GarageFinish: 차고 내부 마감
# GarageCars: 차량 용량 크기
# GarageArea: 평방 피트 단위의 차고 크기
# GarageQual: 차고 품질
# GarageCond: 차고 상태
# PavedDrive: 포장된 진입로
# WoodDeckSF: 목판 면적(평방 피트)
# OpenPorchSF: 개방 현관 면적(평방 피트)
# EnclosedPorch: Enclosed porch area in square feet: 밀폐된 현관 구역(평방 피트)
# 3SsnPorch: Three season porch area in square feet: 3계절 현관 면적(평방 피트)
# ScreenPorch: Screen porch area in square feet: 스크린 현관 면적(평방 피트)
# PoolArea: Pool area in square feet: 수영장 면적(평방 피트)
# PoolQC: Pool quality : 수영장 품질
# Fence: Fence quality : 울타리 품질
# MiscFeature: 기타 기능은 다른 범주에서 다루지 않음
# MiscVal: 기타 기능의 가치
# MoSold: 월 판매
# YrSold: 연도 판매
# SaleType: 판매 유형
# SaleCondition: 판매 조건
```

#### 예측해야 할 칼럼 : SalePrice

```python
import pandas as pd
import numpy as np
```

```python
pd.options.display.max_columns = 999
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
display(train,test)
```

#### 아웃라이어 인덱스 제거

```python
drop_index = train[(train["GrLivArea"] > 4000) & (train["SalePrice"] < 200000)].index
train = train.drop(drop_index)
```

```python
all_data = pd.concat([train,test])
```

```python
all_data.corr()["SalePrice"].sort_values(ascending=False)
```

#### 파생변수

```python
# Yrsold(집이 팔린 년도) - YearBuilt(집이 지어진 년도) = 집이 오래될 수록 싸진다.
all_data["old"] = all_data["YrSold"] - all_data["YearBuilt"]
```

```python
# OverallQual(집의 퀄리티) + OverallCond(집의 컨디션) 
all_data["Overall"] = all_data["OverallQual"] + all_data["OverallCond"]
```

```python
# OverallQual(집의 퀄리티) - OverallCond(집의 컨디션) 
all_data["Overall_minus"] = all_data["OverallQual"] - all_data["OverallCond"]
```

```python
# 집 내부 크기 추가
all_data["total_area"] = all_data["GrLivArea"] + all_data["TotalBsmtSF"]
```

```python
# 집 내부 + 집 외부 , 전체 크기 추가
all_data["final_area"] = all_data["total_area"] + all_data["GarageArea"]
```

```python
# 전체 크기 + "LotArea"
all_data["area"] = all_data["final_area"] + all_data["LotArea"] * 0.1 
# final_area(0~3000) 보다 LotArea(0~20000)의 범위가 훨씬 크기 때문에 final_area의 정보를 잡아먹는다.
```

```python
# 폭과 깊이에 대해서는 "Frontage"가 "Depth"보다 중요하다. 전면이 넓은 땅이 값어치가 넓다.
# 전체 대지에서 "Frontage"의 비율이 높을 수록 집값이 올라갈 것 같다.
all_data["Lot_depth"] = all_data["LotArea"] / all_data["LotFrontage"]
```

```python
# 지하 관련 공사 끝난 경우
all_data["Bsmt"] = all_data["TotalBsmtSF"] - all_data["BsmtUnfSF"]
```

```python
# 차고면적 평균 초과일때, 차 3대 이상 보유
all_data["car"] = np.where((all_data["GarageArea"] > 472) & (all_data["GarageCars"] > 2),1,0)
```

```python
# 현관 + 목판
all_data["Deck_Porch"] = all_data["WoodDeckSF"] + all_data["OpenPorchSF"]
```

```python
# 높은 퀄리티의 면적
all_data["highQualFinSF"] = all_data["GrLivArea"] - all_data["LowQualFinSF"]
```

#### [아웃라이어 제거 후]
그래프는 우상향으로 선형적이다.
GrLivArea"칼럼은 지상의 크기를 의미한다

```python
import warnings
warnings.filterwarnings(action="ignore")
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (16,8))
sns.scatterplot(all_data["GrLivArea"], all_data["SalePrice"])
```

#### [아웃라이어 제거 전]
선과 거리가 멀면서 튀는 두개의 값이 아웃라이어로 확인 되었다.

```python
# 주방 품질
all_data["KitchenQual"] = all_data["KitchenQual"].replace({"Ex":4,"Gd":3,"TA":2,"Fa":1})
```

```python
# 외장재 품질 + 상태 
all_data["ExterQual"] = all_data["ExterQual"].replace({"Ex":4,"Gd":3,"TA":2,"Fa":1})
all_data["ExterCond"] = all_data["ExterCond"].replace({"Ex":4,"Gd":3,"TA":2,"Fa":1,"Po":0})
all_data["Extertotal"] = all_data["ExterQual"] + all_data["ExterCond"] 
```

```python
# 지하실 높이 + 상태
all_data["BsmtQual"] = all_data["BsmtQual"].replace({"Ex":4,"Gd":3,"TA":2,"Fa":1})
all_data["BsmtCond"] = all_data["BsmtCond"].replace({"Gd":3,"TA":2,"Fa":1,"Po":0})
all_data["Bsmttotal"] = all_data["BsmtQual"] + all_data["BsmtCond"]
```

```python
all_data['Utilities'].value_counts() # 한 종류에 데이터가 쏠려 있다. 
```

```python
all_data['Alley'].value_counts() # 결측치가 너무 많다.
```

```python
all_data['MiscFeature'].value_counts() # 결측치가 너무 많다.
```

```python
# 정답값과 불필요한 칼럼 삭제
all_data2 = all_data.drop(columns=["Id","SalePrice","Utilities","Alley","MiscFeature"])
```

```python
# "MSSubClass"(주택 유형) 칼럼을 문자형으로 바꿔주었다.
all_data2["MSSubClass"] = all_data2["MSSubClass"].astype("str")
```

```python
# 추가적으로 아래 두 칼럼을 문자형으로 바꿔보는 실험을 해보았다.
all_data2["OverallQual"] = all_data2["OverallQual"].astype("str")
all_data2["OverallCond"] = all_data2["OverallCond"].astype("str")
```

```python
# 원핫-인코딩
all_data3 = pd.get_dummies(all_data2,drop_first=True)
```

```python
# 결측치 채우기
all_data3 = all_data3.fillna(-1)
```

```python
# 표준화를 통해 칼럼의 범위를 비슷하게 만들어준다.
from sklearn.preprocessing import RobustScaler
rs = RobustScaler()
all_data4 = rs.fit_transform(all_data3)
```

```python
train2 = all_data4[:len(train)]
test2 = all_data4[len(train):]
```

#### y값의 분포가 왼쪽으로 쏠려있다. 오른쪽으로도 튀는 값들이 존재한다.

```python
import seaborn as sns
sns.displot(train["SalePrice"])
```

#### 학습 시 log를 씌우면 x의 범위가 줄어들고, 튀는 값들을 덜 만든다.

```python
sns.displot(np.log(train["SalePrice"]))
```

#### 이 대회는 숫자적 느낌의 칼럼이 매우 우세하다.
하지만 카테고리 칼럼도 많고, 숫자적 느낌의 칼럼도 많기에 각기 잘하는 분야로 학습한다.

#### 앙상블 = 선형 모델 + 트리모델

```python
from sklearn.linear_model import Ridge
rg = Ridge(alpha = 9, random_state = 82)
rg.fit(train2,np.log(train["SalePrice"]))
result = rg.predict(test2)
result
```

```python
from catboost import CatBoostRegressor
cbr = CatBoostRegressor(iterations = 10000,verbose=0) 
cbr.fit(train2,np.log(train["SalePrice"]))
result2 = cbr.predict(test2)
result2
```

#### mse가 Sklearn에 없기 때문에 neg_mean_squared_error를 사용하였다.

```python
from sklearn.model_selection import cross_val_score 
cvs = cross_val_score(rg,train2,train["SalePrice"],n_jobs=4,cv=10,scoring="neg_mean_squared_error")
np.sqrt(-cvs.mean())
```

```python
for i in [5,6,7,8,9,10,11,12,13]:
    rg_cv = Ridge(alpha = i)
    cvs =  cross_val_score(rg_cv,train2,train["SalePrice"],n_jobs=4,cv=10,scoring="neg_mean_squared_error")
    cvs = np.sqrt(-cvs.mean())
    print(f"loss {i} : {cvs}")
```
