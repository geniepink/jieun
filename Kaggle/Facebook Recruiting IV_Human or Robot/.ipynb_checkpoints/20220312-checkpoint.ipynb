```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings(action = "ignore")
```

```python
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
bids = pd.read_csv("bids.csv")
display(train,test,bids)
```

```python
train["outcome"].unique()
# 0이 사람, 1이 로봇
```

```python
time = bids.groupby("bidder_id")["time"].apply(lambda x:(x.max()-x.min()) / x.count())
time
```

```python
from datetime import datetime
bids["datetime"] = bids["time"].apply(lambda x:datetime.fromtimestamp(x/7000000))
```

```python
bids["year"] = bids["datetime"].dt.year
bids["month"] = bids["datetime"].dt.month
bids["day"] = bids["datetime"].dt.day
bids["hour"] = bids["datetime"].dt.hour
```

```python
datetime = bids.groupby(["bidder_id"])["year","month","day"].agg("mean")
datetime
```

```python
# 특정 id가 특정 달에 몇 번 활동했는지에 대한 직접적인 정보를 추가
month_active = bids.groupby(["bidder_id","month"])["bidder_id"].agg(len).unstack()
month_active
```

```python
day_active = bids.groupby(['bidder_id','day'])['bidder_id'].agg(len).unstack().add_prefix("day_")
day_active 
```

```python
hour_active = bids.groupby(["bidder_id","hour"])["bidder_id"].agg(len).unstack().add_prefix("hour_")
```

```python
# bidder_id 하나 당 데이터가 너무 많으니 bidder_id 하나 당 하나의 행만 추가되도록 해야 한다.
# 행동을 몇 번 했는지 알 수 있다.
bids[bids["bidder_id"] == "00486a11dff552c4bd7696265724ff81yeo9v"]
```

```python
bidder_id_count = bids.groupby("bidder_id")["auction"].count().to_frame("bidder_id_count")
bidder_id_count
```

```python
merchandise = bids.groupby("bidder_id")["merchandise"].max()
merchandise
```

```python
nunique = bids.groupby("bidder_id")["auction","device","country"].nunique().add_suffix("_nunique")
# bidder_id 당 auction, device, country 각 종류의 고유 개수
nunique
```

```python
auction_direct = bids.groupby(["bidder_id","auction"])["bidder_id"].agg(len).unstack()
auction_direct
```

```python
bids["auction"].nunique()
```

```python
# train과 test의 종류가 겹치는 것과 안겹치는 것을 구분해야 한다.
train_auction = pd.merge(train,bids[["bidder_id","auction"]], on = "bidder_id", how = "left")
train_auction_list = train_auction["auction"].unique()
train_auction_list
```

```python
test_auction = pd.merge(test,bids[["bidder_id","auction"]], on = "bidder_id", how = "left")
test_auction_list = test_auction["auction"].unique()
test_auction_list
```

```python
auction_interlist = list(set(train_auction_list)&set(test_auction_list))
len(auction_interlist)
```

```python
auction_direct.columns
```

```python
len(auction_direct.columns)
```

```python
auction_direct[auction_direct.columns[auction_direct.columns.isin(auction_interlist)]]
```

```python
country_direct = bids.groupby(["bidder_id","country"])["bidder_id"].agg(len).unstack()
country_direct
```

```python
all_data = pd.concat([train,test])
all_data = all_data.join(bidder_id_count, on = "bidder_id")
all_data = all_data.join(nunique, on = "bidder_id")
all_data = all_data.join(merchandise, on = "bidder_id")
#all_data = all_data.join(auction_direct, on = "bidder_id")
all_data = all_data.join(country_direct, on = 'bidder_id')
all_data = all_data.join(time, on = "bidder_id")
all_data = all_data.join(datetime, on = "bidder_id")
all_data = all_data.join(month_active, on = "bidder_id")
all_data = all_data.join(day_active, on = "bidder_id") 
all_data = all_data.join(hour_active, on = "bidder_id")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(all_data["outcome"],all_data["bidder_id_count"])
# 아웃라이어 확인해보니 오히려 큰 값의 아웃라이어는 사람이 더 많았다.
```

```python
sns.boxplot(all_data["outcome"],all_data["bidder_id_count"], showfliers = False)
```

```python
all_data["auction_ratio"] = all_data["bidder_id_count"] / all_data["auction_nunique"]
# 한 사람이 한 종류의 auction에서 몇 번이나 구매했는지에 대한 비율 
```

```python
all_data2 = all_data.drop(columns=["bidder_id","outcome"])
all_data2
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in all_data2.columns[all_data2.dtypes == "object"]:
    all_data2[i] = le.fit_transform(list(all_data2[i]))
all_data2
```

```python
all_data2 = all_data2.fillna(-1) 
```

```python
train2 = all_data2[:len(train)]
test2 = all_data2[len(train):]
display(train2,test2)
```

```python
from catboost import CatBoostClassifier
cbc = CatBoostClassifier(verbose = 200)
cbc.fit(train2, train['outcome'])
result = cbc.predict_proba(test2)
result
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_jobs = -1, random_state = 82, max_depth = 7, n_estimators = 170)
rf.fit(train2, train["outcome"])
result_rf = rf.predict_proba(test2)
result_rf
```

```python
sub = pd.read_csv("sampleSubmission.csv")
sub["prediction"] = result[:,1] * 0.5 + result_rf[:,1] * 0.5
sub
```

```python
sub.to_csv("sub1.csv",index = False)
```
