```python
action_direct = sess.groupby(["user_id","action"])["user_id"].agg(len).unstack()
action_direct
```

```python
train_action = pd.merge(train,sess[["user_id","action"]].rename(columns={"user_id":"id"}),on = "id", how = "left")
train_action["action"].nunique()
```

```python
test_action = pd.merge(test,sess[["user_id","action"]].rename(columns={"user_id":"id"}), on = "id", how = "left")
test_action["action"].nunique()
```

```python
action_interlist = list(set(train_action["action"]) & set(test_action["action"]))
len(action_interlist) #284개 데이터가 겹친다.
```

```python
action_direct = action_direct[action_interlist[1:]]
action_direct
```

```python
# 추가1
action_type_direct = sess.groupby(["user_id","action_type"])["user_id"].agg(len).unstack().add_prefix("type_")
action_type_direct
```

```python
train_action_type = pd.merge(train,sess[["user_id","action_type"]].rename(columns={"user_id":"id"}),on = "id", how = "left")
train_action_type["action_type"].nunique()
```

```python
test_action_type = pd.merge(test,sess[["user_id","action_type"]].rename(columns={"user_id":"id"}),on = "id", how = "left")
test_action_type["action_type"].nunique()
```

```python
action_type_interlist = list(set(train_action_type["action_type"]) & set(test_action_type["action_type"]))
len(action_type_interlist) 
print(action_type_interlist)
```

```python
action_type_interlist = ['type_' + x for x in action_type_interlist[1:]]
```

```python
action_type_direct = action_type_direct[action_type_direct.columns[action_type_direct.columns.isin(action_type_interlist)]]
action_type_direct
```

```python
action_detail_direct = sess.groupby(["user_id","action_detail"])["user_id"].agg(len).unstack().add_prefix("detail_")
action_detail_direct
```

```python
# 추가2
len(action_detail_direct.columns)

train_action_detail = pd.merge(train,sess[["user_id","action_detail"]].rename(columns = {"user_id":"id"}), on = "id", how = "left")
train_action_detail["action_detail"].nunique()

test_action_detail = pd.merge(test,sess[["user_id","action_detail"]].rename(columns = {"user_id":"id"}), on = "id", how = "left")
test_action_detail["action_detail"].nunique()
```

```python
action_detail_interlist = list(set(train_action_detail["action_detail"].unique()) & set(test_action_detail["action_detail"].unique()))
len(action_detail_interlist)
#print(action_detail_interlist)

action_detail_interlist = ["detail_" + x for x in action_detail_interlist[1:]]

action_detail_direct[action_detail_direct.columns[action_detail_direct.columns.isin(action_detail_interlist)]]
```

```python
# 한 아이디 당 특정 action을 할 때 시간이 총 얼마나 소요되었는가?
second_direct = sess.groupby(['user_id','action'])['secs_elapsed'].agg(sum).unstack().add_prefix("second_")
second_direct
```

```python
second_interlist = ['second_' + x for x in action_interlist[1:]]
```

```python
second_direct = second_direct[second_interlist]
second_direct
```

```python
all_data = pd.concat([train,test])
all_data = all_data.join(id_count,on="id")
all_data = all_data.join(nunique,on="id")
all_data = all_data.join(action_direct,on="id")
all_data = all_data.join(second_direct,on="id")
all_data = all_data.join(action_type_direct,on="id") #0.88303
all_data = all_data.join(action_detail_direct,on="id") # gpu 없애서 test해보기
```

```python
all_data2 = all_data.drop(columns = ["id","country_destination","date_account_created","timestamp_first_active","date_first_booking"])
all_data2
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in all_data2.columns[all_data2.dtypes == "object"]:
    all_data2[i] = le.fit_transform(list(all_data2[i]))
all_data2
```

```python
train2 = all_data2[:len(train)]
test2 = all_data2[len(train):]
display(train2,test2)
```

```python
from catboost import CatBoostClassifier
cbc = CatBoostClassifier(verbose = 200)
cbc.fit(train2, train["country_destination"])
result = cbc.predict_proba(test2)
result
```
